```The file name starts with an 19 digit number, which is the UNIX time in nanoseconds. Using the python datetime library and pytz, convert this number to a python datetime object in both UTC and CERN local time.```



````python
'''Extracting the given date-time stamp from the filename'''


import glob
import datetime
file_name=glob.glob('*.h5')
unixtime=file_name[0].split('_')[0]
````

````python
'''To convert the given time from nano secounds to secounds ,
as the parameters used in datetime only supports an integer 
range of variables'''


unixTimeSecounds=int(float(unixtime)/1000000000)
unixTimedelT=datetime.timedelta(seconds=unixTimeSecounds)
````

````python
'''As the base for the countdown of time for UNIX time starts 
from 00:00 1st JAN , 1970 UTC , 
we add delta to the base time to get the current time'''


UTC_time=datetime.datetime(1970,1,1)+unixTimedelT
print("UTC time :{}".format(UTC_time))
````

````python
'''The Local Cern time is CET / mapping to : Europe/Zurich 
as present in the list of timezones available in timezone()'''


import pytz
timeZ=pytz.timezone('Europe/Zurich')
CERN_time=pytz.utc.localize(UTC_time, is_dst=None).astimezone(timeZ)
print('CERN time(local time):{}'.format(CERN_time))
````

```Use the h5py library to open the hdf file. The file is organized as a directory tree with groups and datasets. Write a program which explores all branches of the directory tree and identifies all of the datasets in the file. Your program should create a csv file which records the names of all of the groups and datasets, and includes the size, shape and type of data in each dataset.```


````python
import h5py as h5
file_toSearch=h5.File(file_name[0],'r')
````

````python
outputCSV=open("output.csv","w+")
outputCSV.write("Group,Dataset,Shape,Data Type,Size\n")
def writeattributes_toCSV(name, obj):
    if(type(obj)==h5._hl.group.Group):
        #print('-'*11+'Group'+'-'*11)
        #print(obj.name)
        outputCSV.write(obj.name+",,,,\n")
        #print('-'*30+'\n')
    else:
        #print('-'*11+'Dataset'+'-'*11)
        #print('Name: '+obj.name)
        try:
            #print('Size: '+str(obj.size))
            #print('Shape: '+str(obj.shape))
            #print('DataType :'+str(obj.dtype))
            outputCSV.write(","+obj.name+","+str(obj.shape)+","+str(obj.dtype)+","+str(obj.size)+"\n")
            
        except TypeError:
            #print('Size: '+str(obj.size))
            #print('Shape: '+str(obj.shape))
            #print('DataType:NaN')
            outputCSV.write(","+obj.name+","+str(obj.shape)+",NaN,"+str(obj.size)+"\n")
        except AttributeError:
            #print('Size:NaN')
            #print('Shape:NaN')
            #print('DataType:'+str(obj.dtype))
            outputCSV.write(","+obj.name+",NaN,"+str(obj.dtype)+",NaN\n")
        
        #print('-'*30+'\n')       
````


The list of groups and datasets as encountered in the walk of the tree , are listed as follows :
(and are written to ```Output.csv```)

````python
file_toSearch.visititems(writeattributes_toCSV)
outputCSV.close()
````

```In the hdf file, there is 2D image data stored as a 1D array in the dataset called "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageData". The datasets "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageHeight" and "/AwakeEventData/XMPP-STREAK/StreakImage/streakImageWidth" store information about the height and width of the image. Use numpy.reshape to convert the 1D array into a 2D image. Then, use scipy.signal.medfilt to filter the image. Finally, use matplotlib to display the image and save it as a png file.```

````python
import numpy as np
from scipy.signal import medfilt
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

#as mentioned in the problem statement : the dataset and the attributes 
#are extracted as from the following location of the .h5 file tree:
dataSetLocation='/AwakeEventData/XMPP-STREAK/StreakImage/streakImageData'
dataSet_heightLocation = '/AwakeEventData/XMPP-STREAK/StreakImage/streakImageHeight'
dataSet_widthLocation = '/AwakeEventData/XMPP-STREAK/StreakImage/streakImageWidth'


#importing the 1D image :
dataSet_width=file_toSearch[dataSet_widthLocation][:][0]
dataSet_height=file_toSearch[dataSet_heightLocation][:][0]
def_1Darray=np.zeros((344064,),dtype='int32')
file_toSearch[dataSetLocation].read_direct(def_1Darray,np.s_[0:dataSet_height*dataSet_width]
                                           ,np.s_[0:dataSet_height*dataSet_width])
````

````python
#reshaping the 1D array to 2D
array2D=np.reshape(def_1Darray,(dataSet_height,dataSet_width))
plt.imshow(array2D)
plt.show()
````

````python
#filtering the image using scipy.signal.medfilt
filteredImage=medfilt(array2D,5)
````

````python
plotFigure = plt.figure()
plt.imshow(filteredImage)
plotFigure.savefig('OutputIMG.png')
````

The filtered image is expoted/saved as ```OutputIMG.png``
